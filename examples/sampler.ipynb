```python
from chembayes import Sampler
```

```python
# Set parameters from dictionary
params = {
    'x1_float': {
        'type': 'float',
        'l_bound': 0.0,
        'u_bound': 1
    },
    'x2_int': {
        'type': 'int',
        'l_bound': 10,
        'u_bound': 100
    },
    'x3_cat': {
        'type': 'categoric',
        'categories': ['A', 'B', 'C'] 
    }
}

sampler = Sampler()
sampler.set_params(params)
sampler.qmc_sampling(20)
```

```python
sampler.get_df()
```

```python
sampler.plot()
```

```python
# Option 2: Set parameters ussing add functions
sampler2 = Sampler()
sampler2.add_float('x1_float', 0.0, 1.0)
sampler2.add_int('x2_int', 10, 100)
sampler2.add_categoric('x3_cat', ['A', 'B', 'C'])

sampler2.qmc_sampling(20)
```

```python
sampler2.get_df()
```

```python
sampler2.plot()
```
