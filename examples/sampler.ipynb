```python
from chembayes import qmc_sampling, plot_qmc_sampling
```

```python
# Define variables, types and bounds
parameters = {
    'x1_float': {
        'type': 'float',
        'l_bound': 0.0,
        'u_bound': 1
    },
    'x2_int': {
        'type': 'int',
        'l_bound': 10,
        'u_bound': 100
    },
    'x3_cat': {
        'type': 'categoric',
        'categories': ['A', 'B', 'C'] 
    }
}
```

```python
# Generate sampling matrix with 20 points
df = qmc_sampling(parameters, 20, force_edges=True)
df
```

```python
# Plot sampling matrix (numeric vars only)
plot_qmc_sampling(df)
```
